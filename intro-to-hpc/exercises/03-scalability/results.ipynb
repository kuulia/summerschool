```python
import pandas as pd
```

```python
df = pd.read_csv('consolidated.tsv', sep='\t')
df['N nodes'] = [1, 2, 4, 8, 16]
df
```

```python
nodes = df['N nodes']
runtime = df['iteration_time_s']
t1 = runtime.iloc[0]

power_per_node_w = 600
price_per_node_hour = 4

results = pd.DataFrame({
    'Nodes': nodes,
    'Runtime (s)': runtime,
})
results['Resource cost (Node-s)'] = results['Nodes'] * results['Runtime (s)']
results['Speedup'] = t1 / results['Runtime (s)']
results['Parallel efficiency (%)'] = 100 * results['Speedup'] / results['Nodes']
results['Electricity cost (Wh)'] = results['Resource cost (Node-s)'] * power_per_node_w / 3600
results['Monetary cost (\u20ac)'] = results['Resource cost (Node-s)'] / 3600 * price_per_node_hour

results
```
