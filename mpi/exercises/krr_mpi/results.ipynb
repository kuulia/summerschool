```python
import pandas as pd
import matplotlib.pyplot as plt

seq  = pd.read_csv("predictions_sequential.txt", sep=" ")
grid = pd.read_csv("predictions_grid.txt",       sep=" ")
```

```python

fig, axes = plt.subplots(1, 2, figsize=(12, 4), sharey=True)
for ax, df, title in zip(axes, [seq, grid], ["Sequential (rank 0→1)", "Grid search (rank 2)"]):
    ax.scatter(df["x1"], df["y_true"],  s=20, label="data",      alpha=0.6)
    ax.scatter(df["x1"], df["y_pred"],  s=20, label="KRR fit",   alpha=0.6)
    ax.set_title(title)
    ax.legend()
plt.tight_layout()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("krr_data.txt", sep=" ", skiprows=1, header=None, names=["x1","y"])
seq  = pd.read_csv("eval_sequential.txt", sep=" ")
grid = pd.read_csv("eval_grid.txt",       sep=" ")

fig, axes = plt.subplots(1, 2, figsize=(13, 6), sharey=True)
for ax, df, title in zip(axes, [seq, grid],
                          ["Sequential (σ_fixed=200, λ=1e-4, σ=0.66)",
                           "Grid search (λ=0.053, σ=0.35)"]):
    ax.scatter(data["x1"], data["y"], s=8, alpha=0.4, label="data")
    ax.plot(df["x1"], df["y_pred"], color="red", lw=1.5, label="KRR fit")
    ax.set_title(title); ax.legend()
plt.tight_layout()
```
